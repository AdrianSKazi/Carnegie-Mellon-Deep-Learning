# Gradient

```python
mat = torch.ones(2, requires_grad = True)
mat
```

```python
A = mat * 2
A
```

```python
print(mat.grad)
```

for now we don't have any output we need to initialize backward, so what changes have been made to a tensor.

```python
A.backward(torch.ones(2))
```

```python
mat.grad
```

```python
mat
```

now we can see how mat was used to change A tensor

```python
x = torch.tensor([1.,2.,3.], requires_grad = True)
y = torch.tensor([1.,2.,3.], requires_grad = True)

x, y
```

```python
F = 3*x**4 - 2*y**2
```

```python
F
```

```python
F.shape
```

We have 3 objects in it so the x, y tensors

```python
grad = torch.ones(3) # we need a placeholder for gradient
```

```python
grad
```

```python
F.backward(grad)
```

the partial derivative of F with respect to x = 12x^3
the partial derivative of F with respect to y = 4y

```python
x.grad
```

```python
y.grad
```

so for example when we use 1st index of x tensor which is 1, in 12x^3 we get:

```python
12*(1**3)
```

which is first instance in x.grad
